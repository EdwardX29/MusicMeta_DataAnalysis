```python
from statsmodels.stats.proportion import proportions_ztest

def checkSig(p):
    print(f"p={p}")
    if p < 0.05:
        print("Significant! p < 0.05 ")
    else:
        print("Not significant p > 0.05 ")

# can we assume anything from our sample
significance = 0.05

# for 2000s music
pos1, neg1, neu1 = 47, 4,3386
total1 = pos1 + neg1 + neu1
# for 2010s music
pos2, neg2, neu2 = 59, 20, 5217
total2 = pos2 + neg2 + neu2

# -------------------- #
print("-"*20)
# print("Testing Positive Proportions")
# test for pos1 > pos2
print("Testing whether 2000 Positive Proportion > 2010 Positive Proportion")
t, p = proportions_ztest(count=[pos1,pos2], nobs=[total1, total2],  alternative='larger')
checkSig(p)

print()
# test for pos2 > pos1
print("Testing whether 2000 Positive Proportion < 2010 Positive Proportion")
t, p = proportions_ztest(count=[pos1,pos2], nobs=[total1, total2],  alternative='smaller')
checkSig(p)

# 1.0565965397263843 0.14534788020275907
# 1.0565965397263843 0.854652119797241

print()

# ------------------------- #
print("-"*20)
# print("Testing Negative Proportions")
# test for neg1 > neg2
print("Testing whether 2000 Negative Proportion > 2010 Negative Proportion")
t, p = proportions_ztest(count=[neg1,neg2], nobs=[total1, total2],  alternative='larger')
checkSig(p)

print()
# test for neg2 > neg1
print("Testing whether 2000 Negative Proportion < 2010 Negative Proportion")
t, p = proportions_ztest(count=[neg1,neg2], nobs=[total1, total2],  alternative='smaller')
checkSig(p)


# (-2.278418656062376, 0.9886491770438874)
# (-2.278418656062376, 0.011350822956112507)

print()

# ------------------------- #
print("-"*20)
# print("Testing Neutral Proportions")
# test for neu1 > neu2
print("Testing whether 2000 Neutral Proportion > 2010 Neutral Proportion")
t, p = proportions_ztest(count=[neu1,neu2], nobs=[total1, total2],  alternative='larger')
checkSig(p)
print()
# test for neu2 > neu1
print("Testing whether 2000 Neutral Proportion < 2010 Neutral Proportion")
t, p = proportions_ztest(count=[neu1,neu2], nobs=[total1, total2],  alternative='smaller')
checkSig(p)
print()

# (0.029555976413352118, 0.4882105878465973)
# (0.029555976413352118, 0.5117894121534027)
# (0.029555976413352118, 0.9764211756931946)
```
