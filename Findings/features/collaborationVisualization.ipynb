```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

songsWFeatures2000=686
totalSongs2000 = 3519   
songsWFeatures2010=1409
totalSongs2010 = 5364   

year = ['2000s','2010s']
features = [songsWFeatures2000, songsWFeatures2010]
total = [totalSongs2000, totalSongs2010]


x = np.arange(len(year))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, features, width, label='With Features', color="royalBlue")
rects2 = ax.bar(x + width/2, total, width, label='Total Songs', color="springgreen")
ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('# of songs')
ax.set_title('Songs with Features per decade')
ax.set_xticks(x, year)
ax.legend()

ax.bar_label(rects1, padding=-1, labels=[f'{x:,.0f}' for x in rects1.datavalues])
ax.bar_label(rects2, padding=-1, labels=[f'{x:,.0f}' for x in rects2.datavalues])

fig.tight_layout()

plt.show()
```
