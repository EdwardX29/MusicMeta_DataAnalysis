```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

total2000=1042614     
unique2000=61706        
total2010=1888422     
unique2010=112322     

year = ['2000s','2010s']
unique = [unique2000, unique2010]
total = [total2000, total2010]


x = np.arange(len(year))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, unique, width, label='Unique words')
rects2 = ax.bar(x + width/2, total, width, label='Total words')
ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('# of words')
ax.set_title('Words per decade')
ax.set_xticks(x, year)
ax.legend()

ax.bar_label(rects1, padding=-1, labels=[f'{x:,.0f}' for x in rects1.datavalues])
ax.bar_label(rects2, padding=-1, labels=[f'{x:,.0f}' for x in rects2.datavalues])

fig.tight_layout()

plt.show()
```
