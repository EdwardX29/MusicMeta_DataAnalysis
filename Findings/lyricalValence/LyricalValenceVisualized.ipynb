```python
'''
2000s
47      (positive)
4       (negative)
3386    (neutral)
------------------
2010s
59      (positive)
20      (negative)
5217    (neutral)
'''

import matplotlib.pyplot as plt
import numpy as np


year = ['2000s','2010s']
pos = [47, 59]
neg = [4, 20]
neu = [3386, 5217]


x = np.arange(len(year))  # the label locations
width = 0.8  # the width of the bars

fig, ax = plt.subplots()
# rects1 = ax.bar(x - width/2, pos, width, label='Positive')
rects1 = ax.bar(x-width/3, pos, width/3, label='Positive', color="b")
rects2 = ax.bar(x, neg, width/3, label='Negative', color="r")
rects3 = ax.bar(x + width/3, neu, width/3, label='Neutral', color="g")


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('# of songs')
ax.set_title('Songs by Lyrical Valence')
ax.set_xticks(x, year)
ax.legend()

ax.bar_label(rects1, padding=-1)
ax.bar_label(rects2, padding=-1)
ax.bar_label(rects3, padding=-1)


fig.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


year = ['2000s','2010s']
pos = [47, 59]
neg = [4, 20]


x = np.arange(len(year))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, pos, width, label='Positive',color="b")
rects2 = ax.bar(x + width/2, neg, width, label='Negative', color="r")

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('# of songs')
ax.set_title('Positive & Negative Songs by Lyrical Valence')
ax.set_xticks(x, year)
ax.legend()

ax.bar_label(rects1, padding=-1)
ax.bar_label(rects2, padding=-1)

fig.tight_layout()

plt.show()
```
