```python
# End Class Definition #
# ------------------------------------- # 
# Instantiate List and Dict names for Analysis #
from analyse import Analyse
from scipy.stats import *

listNames = [
  "Acoustic",
  "dance",
  "energy",
  "Instrumental",
  "Live",
  "loud",
  "Speech",
  "Tempo",
  "Valence",
  "duration"
]

dictNames = [
  "mode",
  "key",
  "time",
]

# ----------------------------- #
# Actually Analyse data #

a = Analyse("2000")
b = Analyse("2010")

print(a.getTotalSongs())
print(a.getTotalFeatures())

print(b.getTotalSongs())
print(b.getTotalFeatures())

print()

for name in listNames:
  print(name)
  aList = a.getList(name)
  bList = b.getList(name)
  print(len(aList))
  print(len(bList))
  
  print("Averages: ")
  print(a.getAvgList(aList))
  print(b.getAvgList(bList))
  print("Medians: ")
  print(a.getMedianList(aList))
  print(b.getMedianList(bList))
  print("IQRs: ")
  print(a.getIQRList(aList))
  print(b.getIQRList(bList))
  print("STDs: ")
  print(a.getSTDList(aList))
  print(b.getSTDList(bList))
  TwoTail = ttest_ind(aList, bList, equal_var=True)
  if TwoTail.pvalue < 0.05: print("significant") 
  else: print("not significant")
  print(TwoTail.pvalue)
  
  print()

for name in dictNames:
  aDict = a.getDict(name)
  bDict = b.getDict(name)  
  print(a.getFrequenciesDict(a.getDict(name)))
  print(b.getFrequenciesDict(b.getDict(name)))
```
